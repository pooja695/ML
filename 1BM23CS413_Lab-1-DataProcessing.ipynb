```python
#To display information of all columns
df
```

```python
#Handling Missing Values
df.isnull().sum()
```

```python
#Handling Categorical Data
print("Handling Categorical Data...")
df = pd.get_dummies(df, drop_first=True)
df
```

```python
#Handling Outliers using IQR method (only for numerical columns)
print("Handling Outliers...")
numeric_cols = df.select_dtypes(include=[np.number]).columns  # Select only numeric columns
```

```python
Q1 = df[numeric_cols].quantile(0.25)
Q1
```

```python
Q3 = df[numeric_cols].quantile(0.75)
Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
lower_bound = Q1 - 1.5 * IQR
lower_bound
```

```python
upper_bound = Q3 + 1.5 * IQR
upper_bound
```

```python
df = df[~((df[numeric_cols] < lower_bound) | (df[numeric_cols] > upper_bound)).any(axis=1)]
```
